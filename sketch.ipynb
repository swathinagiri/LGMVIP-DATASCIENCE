```python
import cv2 as cv
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
img_path='/kaggle/input/korean-band-bts-members-face-recognition/images/jimin/jimin6.png'
```

```python
img=cv.imread(img_path)
jimin_image=plt.imshow(img)
plt.show()
```

```python
b,g,r=img[0, 0]
```

```python
print(b,g,r)
```

```python
gray_image = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
```

```python
gray_img=plt.imshow(gray_image,cmap='gray')
plt.show()
```

```python
inverted_image = 255 - gray_image
```

```python
inv_img=plt.imshow(inverted_image,cmap='gray')
plt.show()
```

```python
blurred = cv.GaussianBlur(inverted_image, (21, 21), 0)
plt.imshow(blurred,cmap='gray')
plt.show()
```

```python
inverted_blurred = 255 - blurred
plt.imshow(inverted_blurred,cmap='gray')
plt.show()
```

```python
pencil_sketch = cv.divide(gray_image, inverted_blurred,scale=256.0)
```

```python
plt.imshow(pencil_sketch,cmap='gray')
plt.title("Pencil sketch of park jimin")
plt.show()
```
